```python
import numpy as np
import matplotlib.pyplot as plt
```

# ①点の準備

```python
n=11
```

```python
f = lambda x: -1 + (2/(1 + 16 * x ** 2))
```

```python
#ちなみにこれは以下と一緒
def f(x):
    return -1 + (2/(1 + 16 * x ** 2))
```

```python
x = np.linspace(-1, 1, n)#-1~1の　間に等間隔で11個のベクトルを生成
x
```

```python
y = f(x)
y
```

# ②点から多項式回帰モデルの係数を決定する

```python
#点から多項式回帰モデルを最小二乗法し係数を割り出している。ラグランジュ補間という。
#ラグランジュ補間では点の数-1の次数の多項式回帰モデルでのフィティングを行っていることになる。
#この場合は、11個の点から1引いた10次の多項式回帰モデルの係数を求めている
p = np.polyfit(x, y, n-1)
p
```

# ③求めた多項式回帰モデル上の200個の点を生成

```python
xx = np.linspace(-1, 1, 200)#-1~1の　間に等間隔で200個のベクトルを生成
xx;
```

```python
#-278xの11乗+629xの9乗-494xの7乗+167xの5乗-25xの3乗+xの1乗このモデルに200個のxxを代入してyyを出す
yy = np.polyval(p, xx)#pという係数を持つ多項式回帰モデルにxを当てはめてyyをもとめよ
yy;
```

# ④最後にplotしていく

```python
plt.plot(x, y, 'ro') #11打点についてプロット
```

```python
plt.plot(x, y)#ちなみに3つ目の引数がないと折れ線になって表示される
```

```python
plt.plot(xx, yy, 'm') #200個の点が折れ線で繋がれて曲線に。
```

```python
plt.plot(xx, f(xx))#最初に与えた関数のプロット
plt.xlabel('x')
plt.ylabel('y')
plt.title('-1 + (2/(1 + 16 * x ** 2))')
```

```python
#青→点を生み出すために使った最初の関数
#赤→最初の関数から生み出した11個の点
#ピンク→11個の点をフィッティングして作った多項式回帰モデル
plt.plot(x, y, 'ro') 
plt.plot(xx, yy, 'm') 
plt.plot(xx, f(xx)) #Original function
plt.xlabel('x')
plt.ylabel('y')
plt.title('-1 + (2/(1 + 16 * x ** 2))')
```

あれ，青い曲線(実際の関数)に合わないぞ・・・

それもそのはずです．このスクリプトだと11点に対して10次式補間しているため，存在する点から表せる10次式を出しているに過ぎません．これを「発散する」と言います．

これは多項式補間のデメリットなのですが，等間隔のxの数値に対しては実際の関数からはかけ離れた関数の数値を出しやすいという特徴があります．

それでは，次はこの発散を止めるためにデータの間隔を変えてみましょう．

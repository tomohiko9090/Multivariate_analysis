```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 11
```

```python
f = lambda x: -1 + (2/(1 + 16 * x ** 2))
```

```python
x = -1 * np.cos(-1 * np.pi * np.arange(0, n)/(n-1))#xを使って端っこの方は細かく値をとるようにする
x
```

# 以下は多項式補間(失敗と同様)

```python
y = f(x)
y
```

```python
p = np.polyfit(x, y, n-1)
xx = np.linspace(-1, 1, 200)
yy = np.polyval(p, xx)
```

```python
#Plot data
plt.plot(x, y, 'ro') #Sampled data
#plt.hold(True)
plt.plot(xx, yy, 'm') #Polynominal interpolation
#plt.hold(True)
plt.plot(xx, f(xx)) #Original function
plt.xlabel('x')
plt.title('-1 + (2/(1 + 16 * x ** 2))')
plt.savefig('8_1_b.png', dpi=1200)
```

端の方も補間できていることがわかる
